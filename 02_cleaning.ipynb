```python
import json

def clean_data(data):
    
    # Keep only users whose name is not empty or spaces and remove user who have not friends and liked_pages
    data['users'] = [
        user for user in data['users']
        if user['name'].strip() and user['friends'] or user['liked_pages']
    ]
    # Remove the Duplicates Friends
    for user in data['users']:
        user['friends'] = list(set(user['friends']))

    unique_pages = {}
    for page in data['pages']:
        unique_pages[page['id']] = page
    data['pages'] = list(unique_pages.values())

        
    return data

   


# Load the data
with open("data2.json", "r") as f:
    data = json.load(f)

# Clean the data
data = clean_data(data)

# Save cleaned data
with open("cleaned_data2.json", "w") as f:
    json.dump(data, f, indent=4)
```
