```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("data.json")
```

```python
def display(data):
    print("User and their connections\n")

    # create id â†’ name map
    user_map = {}
    for u in data["users"]:
        user_map[u["id"]] = u["name"]

    for user in data["users"]:
        friend_names = []
        for fid in user["friends"]:
            friend_names.append(user_map[fid])

        print(
            f"Id-{user['id']} {user['name']} is friend with "
            f"{', '.join(friend_names)} and liked pages are {user['liked_pages']}"
        )

    print("\nPages Information\n")
    for page in data["pages"]:
        print(f"{page['id']}: {page['name']}")
```
